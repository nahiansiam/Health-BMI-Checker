```python
# 🏥 Health BMI Checker
# Created with Python
# Created by Nahian Kabir Siam

def get_bmi(weight, height):
    return round(weight / (height * height), 2)


def health_status(bmi):
    if bmi < 18.5:
        return "🟡 Underweight"
    elif bmi < 25:
        return "🟢 Healthy Weight"
    elif bmi < 30:
        return "🟠 Overweight"
    else:
        return "🔴 Obese"


print("\n" + "═" * 45)
print("🏥      HEALTH BMI CHECKER      🏥")
print("═" * 45)

try:
    print("\n📥 Enter Your Information")

    weight = float(input("⚖️ Weight (kg) : "))
    height_cm = float(input("📏 Height (cm) : "))

    if weight <= 0 or height_cm <= 0:
        print("\n❌ Invalid value detected!")
        exit()

    # Convert cm to meter
    height_m = height_cm / 100

    bmi = get_bmi(weight, height_m)
    status = health_status(bmi)

    print("\n" + "─" * 35)
    print("📊 YOUR BMI REPORT")
    print("─" * 35)

    print(f"⚖️ Weight      : {weight} kg")
    print(f"📏 Height      : {height_cm} cm")
    print(f"🧮 BMI Score   : {bmi}")
    print(f"❤️ Health Type : {status}")

    print("─" * 35)

    if bmi < 18.5:
        print("💡 Tip: Try to gain some healthy weight.")
    elif bmi < 25:
        print("🎉 Great! Your BMI is in a healthy range.")
    elif bmi < 30:
        print("🚶 Tip: Regular exercise can help improve your BMI.")
    else:
        print("🥗 Tip: Consider a balanced diet and active lifestyle.")

    print("\n✅ Calculation Completed Successfully!")

except ValueError:
    print("\n❌ Please enter numbers only!")
```
